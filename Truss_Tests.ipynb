```python
from Truss_2D import Truss_2D
```

```python
nodes = {1:[0,0], 2:[4,0], 3:[2,3]}
elements = {1: [2,1], 2:[3,1], 3:[3,2]}
supports = {1:[1,1], 2:[0,1]}
forces = {3:[5,-10]}
elasticity = {x: 210e6 for x in elements}
cross_area = {x: 1e-4 for x in elements}
```

```python
Truss_1 = Truss_2D(nodes = nodes,
                  elements = elements,
                  supports = supports,
                  forces = forces,
                  elasticity = elasticity,
                  cross_area = cross_area)
```

```python
Truss_1.Draw_Truss_Setup(figure_size = [10,8], length_of_arrow=0.5)
```

```python
Truss_1.Solve()
```
